```python
import psutil
import time
```

```python
print(f'Monitroing cpu usage...{psutil.cpu_percent()}')
```

```python
def cpu_monitoring(threshold):
    try:
        while True:
            #check for cpu usage
            cpu_usage = psutil.cpu_percent()
            print(f'Monitoring CPU usage...\n')

            #if usage > threshold alert
            if cpu_usage>threshold:
                print(f'Alert! CPU usage exceeds threshold: {cpu_usage}%\n')

            
            time.sleep(10)
    #exception handling for keyboard interupt      
    except KeyboardInterrupt:
        print('Monitoring Stopped\n')
    #exception handling for any other errors at runtime
    except Exception as e:
        print(f'An error has occurred: {str(e)}')
```

```python
if __name__=='__main__':
#     define your required thresold here
      threshold=80
      cpu_monitoring(threshold)
 
```
